**Preparing the dataset**

```python
# to convert the dataset into the dataframe
# dataset = pd.DataFrame(housing.data)

# to conver the dataset into the dataframe with column names
dataset = pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
dataset["Price"]=housing.target
```

```python
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('/kaggle/working/scaling.pkl','wb'))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
model = lr.fit(X_train,Y_train)
```

```python
# to get the all parameters in which the model is trained
# these are the values used inside the lr.fit(---)
model.get_params()
```

```python
# to print the all points (x1,y1) coefficients in the regression line
print(model.coef_)
```

```python
# to print the intecepts at the axis
model.intercept_
```

```python
reg_pred = model.predict(X_test)
```

```python
plt.scatter(Y_test,reg_pred)
plt.plot
```

```python
residual = Y_test - reg_pred
```

```python
residual
```

```python
sns.displot(residual,kind="kde")
```

```python
plt.scatter(reg_pred,residual)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print("mean squared root error",mean_squared_error(Y_test,reg_pred))
print("mean squared absolute error",mean_absolute_error(Y_test,reg_pred))
print("mean squared root error",np.sqrt(mean_squared_error(Y_test,reg_pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(Y_test, reg_pred)
print(score)
```

```python
1- (1-score)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)
```
